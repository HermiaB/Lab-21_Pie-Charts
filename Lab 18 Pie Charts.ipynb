### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here

import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Drop rows with missing data in 'DatabaseWantToWorkWith'
df_db = df.dropna(subset=['DatabaseWantToWorkWith'])

# Step 2: Split the semicolon-separated list of databases into individual entries
databases_series = df_db['DatabaseWantToWorkWith'].str.split(';').explode()

# Step 3: Count occurrences of each database
db_counts = databases_series.value_counts()

# Step 4: Select top 5 databases
top5_dbs = db_counts.head(5)

# Step 5: Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top5_dbs.values, 
    labels=top5_dbs.index, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=plt.cm.Paired.colors
)
plt.title('Top 5 Databases Respondents Want to Work With')
plt.axis('equal')  # Equal aspect ratio ensures pie chart is circular.
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here

# Step 1: Drop rows with missing 'DevType'
df_dev = df.dropna(subset=['DevType'])

# Step 2: Split the semicolon-separated list into individual developer types
dev_types_series = df_dev['DevType'].str.split(';').explode()

# Step 3: Count occurrences of each developer type
dev_type_counts = dev_types_series.value_counts()

# Step 4: Select top 5 developer types
top5_dev_types = dev_type_counts.head(5)

# Step 5: Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top5_dev_types.values,
    labels=top5_dev_types.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Set3.colors
)
plt.title('Top 5 Developer Roles')
plt.axis('equal')  # Equal aspect ratio to make the pie circular
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
##Write your code here

# Step 1: Drop rows with missing 'OpSysProfessional use'
df_os = df.dropna(subset=['OpSysProfessional use'])

# Step 2: Split the semicolon-separated list into individual OS entries
os_series = df_os['OpSysProfessional use'].str.split(';').explode()

# Step 3: Count occurrences of each OS
os_counts = os_series.value_counts()

# Step 4: Select top OS (you can decide how many to show, e.g., top 5)
top_os = os_counts.head(5)

# Step 5: Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top_os.values,
    labels=top_os.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Pastel1.colors
)
plt.title('Top Operating Systems Used Professionally')
plt.axis('equal')  # Equal aspect ratio ensures pie is a circle
plt.show()
```

### Task 2: Additional Visualizations and Comparisons


##### 2.1 Pie Chart for Top 5 Programming Languages Respondents Have Worked With


The `LanguageHaveWorkedWith` column contains the programming languages that respondents have experience with. We’ll plot a pie chart to display the composition of the top 5 languages.



```python
##Write your code here

# Step 1: Drop rows where 'LanguageHaveWorkedWith' is missing
df_lang = df.dropna(subset=['LanguageHaveWorkedWith'])

# Step 2: Split the semicolon-separated list into individual languages and explode into rows
languages_series = df_lang['LanguageHaveWorkedWith'].str.split(';').explode()

# Step 3: Count frequencies of each language
language_counts = languages_series.value_counts()

# Step 4: Select top 5 languages
top_languages = language_counts.head(5)

# Step 5: Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top_languages.values,
    labels=top_languages.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Set3.colors
)
plt.title('Top 5 Programming Languages Respondents Have Worked With')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

##### 2.2 Pie Chart for Top Collaboration Tools used in Professional Use


Using the `NEWCollabToolsHaveWorkedWith` column, we’ll identify and visualize the top collaboration tools respondents use in their professional work.



```python
##Write your code here

# Step 1: Drop rows with missing values in the column
df_tools = df.dropna(subset=['NEWCollabToolsHaveWorkedWith'])

# Step 2: Split the semicolon-separated list and explode to get one tool per row
tools_series = df_tools['NEWCollabToolsHaveWorkedWith'].str.split(';').explode()

# Step 3: Count frequencies of each tool
tool_counts = tools_series.value_counts()

# Step 4: Select top 5 collaboration tools
top_tools = tool_counts.head(5)

# Step 5: Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top_tools.values,
    labels=top_tools.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Paired.colors
)
plt.title('Top 5 Collaboration Tools Used Professionally')
plt.axis('equal')  # Equal aspect ratio to make pie circular
plt.show()
```

##### 3.1 Pie Chart of `Respondents` Most Admired Programming Languages
